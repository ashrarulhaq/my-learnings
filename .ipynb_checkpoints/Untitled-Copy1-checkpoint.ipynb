```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
import matplotlib.pyplot as plot
```

```python
data = pd.read_csv("Spotify-2000.xls")
data.head()
```

```python
plot.scatter(data["Year"], data["Top Genre"], data['Popularity'], color="yellow")

plot.xlabel("Year")
plot.ylabel("Top Genre")
plot.zlabel('Popularity')
plot.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a figure and a 3D Axes object
fig = plt.figure()
ax = fig.add_subplot(113, projection='3d')

# Generate some data
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]
z = [0, 1, 2, 3, 4]

# Plot the data
ax.scatter(x, y, z)

# Set the x, y, and z labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# Show the plot
plt.show()
```
