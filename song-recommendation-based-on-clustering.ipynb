### Creating a Clustering Algorithm to classify songs

```python
db = pd.read_csv('/kaggle/input/dataset-of-songs-in-spotify/genres_v2.csv')
```

#### Merging all songs_names and playlist into 1

```python
db['song_name'] = db['song_name'].fillna(db['title'])
db['song_name'] = db['song_name'].fillna(db['Unnamed: 0'])
db = db.drop(['Unnamed: 0','title'],axis= 1)
db = db.dropna()
```

#### Scaling Data to cluster

```python
db.describe()
```

Parameters to scale: danceability, energy, key, loudness, mode, speachiness, acousticness, instrumentalness, liveness, valence, tempo, duration_ms, time_signature

We use a Standard Scaler to scale values between (-1,1)

```python
scaler = StandardScaler()
db[['danceability', 'energy','key' , 'loudness' , 'mode' , 'speechiness','acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms', 'time_signature']] = scaler.fit_transform(db[['danceability', 'energy','key' , 'loudness' , 'mode' , 'speechiness','acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms', 'time_signature']])
```

```python
n_clusters = 15
```

```python
db.head()
```

Preparig data for clustering

```python
cluster_data = db.drop(['genre','song_name'],axis=1)
```

### K-means clustering

```python
kmeans = KMeans(n_clusters, random_state=42,max_iter=300)
kmeans.fit(cluster_data)
```

**Replaceing Genres to categories to compare with clustering results**

```python
db_cat_genres = db.replace(db['genre'].unique(),[i for i in range(n_clusters)])
```

```python
clusters = kmeans.labels_
```

```python
plt.figure(figsize = (10,10))
sns.scatterplot(x = db['loudness'],y= db['energy'], hue=clusters)
```

```python
Counter(clusters)
```

```python
cluster_genres = {}
for i in set(clusters):
    array = np.array(clusters)
    indices = np.where(array == i)[0]
    cluster_genres[i] = pd.Series(list(db.iloc[list(indices)]['genre']))
```

```python
plt.figure(figsize = (20,80))
for i in range(len(cluster_genres)):
    plt.subplot(8,2,i+1)
    keys = ['Most Appeared Genres in Given Cluster','Other']
    most = cluster_genres[i].value_counts()[0]
    other = len(cluster_genres[i]) - most
    palette_color = sns.color_palette('bright')
    plt.pie([most, other], labels=keys, colors=palette_color, autopct='%.0f%%')
    plt.title(cluster_genres[i].value_counts().index[0])
plt.show()
```

### Insights
* We can see that K-means failed to cluster the songs accurately, therefore we will try to clutser the songs into less clusters and use other clustering algorithms
